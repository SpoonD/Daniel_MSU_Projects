## Discussion and Conclusion

The light curve of the type 1a supernova shows that what is left over is a neutron star. This means that the star began to expand and brighten up over time. After the expanding was over it began to shed its outer layers, slowly losing its brightness as seen in the light curve. Once this was over you are left with a somewhat dim neutron star. Which has a high magnitude, corresponding to slightly dim. It began at an apparent magnitude of 18 and ended around 21.

With a very similar light curve to the type 1a supernova, we know that it does not become a black hole. The difference here is that a type 1b supernova lacks hydrogen in its outer layer, which is why it does not take as long to finish going supernova. This supernova could not be the sun because our sun has plenty of hydrogen and the hydrogen contained within our sun is so much it will not just vanish or go away that quickly.

In this light curve, we see that the object immediately starts off very bright, with an apparent magnitude of 12 or so. As time goes on, it starts to take a clear decline on its magnitude. This means it is going dim. Leading to the idea that this is either becoming a neutron star, which is extremely dim, or it turns into a black hole, only being visible or detectable visually by seeing matter get sucked into it. We also determined that in order to ever turn into a black hole, the original star must be quite massive, about 8 times the mass of the Sun. We can remove this supernova from our prediction of how our Sun will die. 

Since it cannot be type 1b, or type 2 supernovae, the way our star is most likely to go out will be a type 1a supernova, eventually becoming a neutron, or possibly a planetary nebula. This answers our question of what type of death will our Sun eventually have.
